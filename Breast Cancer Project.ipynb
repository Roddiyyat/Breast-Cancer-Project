Tidying data. What does it mean to tidy a dataset?
Tyding a dataset is important especially when building a statistical of ML model that can be used prediction. 

Occassionaly, we get dataset that we can start analysing immediately, but this isn't always the case; sometime we get messy data that need cleaning.

```python
# messy data that needs cleaning
my_df2 = pd.read_csv('TCGA_genetic_data.csv')
```

```python
my_df2.head()
```

```python
my_df2.info
```

### We can classify our dataset as messy because variables are being stored in rows. One of the functions of tidying our data is to reduce the column variable; this can be done by melting.

```python
# based on the df info output, its safe to assume that our data is a wide data(has lots of columns) 
# we need to convert from wide data(lots of columns) to long data(lots of rows)

#using the pandas melt functtion
# id_vars is for the column we don't want to change
TCGA_long = pd.melt(my_df2, id_vars='bcr_patient_barcode')

TCGA_long.head()
```

```python
TCGA_long = pd.melt(my_df2, id_vars='bcr_patient_barcode',
                    var_name='bcr_biopsy_barcode')

TCGA_long.head()
```

```python
my_df2_melt = pd.melt(my_df2,
    id_vars=['bcr_patient_barcode','bcr_biopsy_barcode'],
    var_name='variables',
    value_name='gen_data',
)
```

```python
my_df2_melt.head()
```

```python
# splitting rows in our dataset

'TCGA-EK-A2H0-01A-11R-A180-07'.split('_')
```

```python
bcr_biopsy_barcode_split = my_df2['bcr_biopsy_barcode'].str.split('_')

bcr_biopsy_barcode_split.head()
```

```python
# now that we've tidyed our data, the next step is to create a new csv file for our dataset

my_df2_melt.to_csv('my_df2.csv', index=False)
```
